```python
import requests
import json
from flask import Flask, jsonify
import pandas as pd
from operator import itemgetter

input_list = ['YYYY','Q','상권_구분','상권_구분명','상권','상권명','업종_코드','업종명','분기당_sales','분기당_sales_count','주중_sales','주말_sales','점포수']
df_sales_2021_1 = pd.DataFrame(columns=input_list)

start = [i for i in range(1, 40000, 1000)]
end = [i for i in range(1000, 40001, 1000)]

for i, j in zip(start, end):
    API_URL = f"http://openapi.seoul.go.kr:8088/76485a52596d616e343153594f6e79/json/VwsmTrdarSelngQq/{i}/{j}/2021"
    raw_data = requests.get(API_URL)
    df_raw = json.loads(raw_data.text)
    length = len(df_raw['VwsmTrdarSelngQq']['row'])
    item = itemgetter('STDR_YY_CD', 'STDR_QU_CD', 'TRDAR_SE_CD', 'TRDAR_SE_CD_NM', 'TRDAR_CD','TRDAR_CD_NM', 'SVC_INDUTY_CD', 'SVC_INDUTY_CD_NM', 'THSMON_SELNG_AMT', 'THSMON_SELNG_CO', 'MDWK_SELNG_AMT', 'WKEND_SELNG_AMT', 'STOR_CO')
    df_raw2 = df_raw['VwsmTrdarSelngQq']['row']
    
    index = len(df_sales_2021_1)
    for k in range(length):
        df_sales_2021_1.loc[index+k] = item(df_raw2[k])

df_sales_2021_1
```

```python
df_sales_2021_1.Q.value_counts()
```
