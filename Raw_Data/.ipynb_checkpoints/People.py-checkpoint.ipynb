```python
import requests
import json
from flask import Flask, jsonify
import pandas as pd
from operator import itemgetter
import pickle

input_list = ['YYYY','Q','상권_구분','상권_구분명','상권','상권명','총_생활인구','남성_생활인구', '여성_생활인구']
df_living_2021 = pd.DataFrame(columns=input_list)


start = [i for i in range(1, 3000, 1000)]
end = [i for i in range(1000, 3001, 1000)]

for i, j in zip(start, end):
    API_URL = f"http://openapi.seoul.go.kr:8088/76485a52596d616e343153594f6e79/json/VwsmTrdarFlpopQq/{i}/{j}/2021"
    raw_data = requests.get(API_URL)
    df_raw = json.loads(raw_data.text)
    length = len(df_raw['VwsmTrdarFlpopQq']['row'])
    item = itemgetter('STDR_YY_CD', 'STDR_QU_CD', 'TRDAR_SE_CD', 'TRDAR_SE_CD_NM', 'TRDAR_CD','TRDAR_CD_NM', 'TOT_FLPOP_CO', 'ML_FLPOP_CO', 'FML_FLPOP_CO')
    df_raw2 = df_raw['VwsmTrdarFlpopQq']['row']

    num = len(df_living_2021)
    for k in range(length):
        df_living_2021.loc[num+k] = item(df_raw2[k])

df_living_2021
```
